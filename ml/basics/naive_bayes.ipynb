## 数学原理

**贝叶斯定理 + 特征独立假设：**

$$P(y \mid x) \propto P(y) \prod_{i=1}^d P(x_i \mid y)$$

**高斯似然：**

$$P(x_i \mid y) = \frac{1}{\sqrt{2\pi\sigma_{iy}^2}} \exp\left(-\frac{(x_i - \mu_{iy})^2}{2\sigma_{iy}^2}\right)$$
