## 数学原理

**欧氏距离：**

$$d(x, y) = \sqrt{\sum_{i=1}^d (x_i - y_i)^2}$$

**预测（多数投票）：**

$$\hat{y} = \text{majority vote of } k \text{ nearest neighbors}$$
